```
from abc import ABC, abstractmethod

# 1. ABSTRACTION
class BankAccount(ABC):
    @abstractmethod
    def transaction(self):
        pass


# 2. ENCAPSULATION
class Account(BankAccount):
    def __init__(self, name, acc_no, balance):
        self.name = name
        self._acc_no = acc_no
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def update_balance(self, amount):
        self.__balance += amount

    def transaction(self):
        print("Basic bank transaction recorded.")


# 3. INHERITANCE
class ATM(Account):

    # 4. CLASS METHOD
    @classmethod
    def bank_info(cls):
        print("\n--- WELCOME TO ABC BANK ATM ---")

    # 5. STATIC METHOD
    @staticmethod
    def check_amount(amt):
        return amt > 0

    # 6. POLYMORPHISM (method override – SAME signature)
    def transaction(self):
        print("ATM transaction processed.")

    # 7. INSTANCE METHODS
    def deposit(self, amt):
        if self.check_amount(amt):
            self.update_balance(amt)
            self.transaction()
            print("Deposit Successful!")
        else:
            print("Invalid amount.")

    def withdraw(self, amt):
        if self.check_amount(amt) and amt <= self.get_balance():
            self.update_balance(-amt)
            self.transaction()
            print("Please collect your cash.")
        else:
            print("Invalid amount or insufficient funds.")


# 8. MENU
def run_atm():
    my_atm = ATM("jhansi", 29007, 1000)
    ATM.bank_info()

    while True:
        print("\n1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Account Details")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            print(f"Balance: ₹{my_atm.get_balance()}")

        elif choice == '2':
            amt = int(input("Amount to deposit: "))
            my_atm.deposit(amt)

        elif choice == '3':
            amt = int(input("Amount to withdraw: "))
            my_atm.withdraw(amt)

        elif choice == '4':
            print(f"Holder: {my_atm.name}")
            print(f"Account No: {my_atm._acc_no}")

        elif choice == '5':
            print("Goodbye!")
            break

        else:
            print("Wrong choice, try again.")


# START
run_atm()
```
